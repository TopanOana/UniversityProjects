Problem: Weighted N-Queen Problem


You are given an N×N chessboard, and each cell of the board has an associated weight. Your task is to find a valid placement of N queens such that the total weight of the queens is maximized, and no two queens threaten each other.





In the traditional N-Queen Problem, the goal is to place N queens on an N×N chessboard in such a way that no two queens threaten each other. In this variation, we introduce weights to the queens and aim to find a placement that maximizes the total weight of the queens while satisfying the constraint of non-threatening positions.


Constraints:

1. There should be exactly one queen in each row and each column.
2. No two queens should be placed in the same diagonal, i.e., they should not threaten each other.
3. The placement should maximize the total weight of the queens.


Representation:

Use a permutation-based representation. Each permutation represents the column position of the queen for each row. 

For example, if N=4, a valid permutation [2, 4, 1, 3] indicates that the queen in the first row is placed in column 2, the queen in the second row is placed in column 4, and so on.


Genetic Algorithm Steps:

1. *Initialization*: Generate an initial population of permutations randomly.

2. *Fitness Evaluation*: Evaluate the fitness of each permutation by calculating the total weight of the queens while considering the non-threatening positions.

3. *Selection*: Select a subset of permutations from the population based on their fitness, using selection techniques like tournament selection or roulette wheel selection.

4. *Crossover*: Perform crossover (recombination) on the selected permutations to create new offspring permutations.

5. *Mutation*: Introduce random changes (mutations) in the offspring permutations to maintain diversity in the population.

6. *Fitness Evaluation for the new individuals*: Evaluate the fitness of the new population.

7. *Form the new population*: Select the surviving individuals based on scores, with chances direct proportional with their performance.

8. Repeat steps 3-7 for a certain number of generations or until a termination condition is met (e.g., a maximum number of iterations or a satisfactory solution is found).


9. *Termination*: Return the best-performing individual (permutation) found as the solution to the problem.

Note: The fitness function used in this problem should calculate the total weight of the queens based on the positions specified by the permutation. Additionally, the fitness function should penalize solutions that violate the non-threatening constraint by assigning a lower fitness score to such permutations.

```python
#N queens problem but weighted
#sounds spicy but fun lesgo
import random
import math

#just going to test with 8
N = 8
weights = [
    [1, 3, 2, 4, 2, 3, 1, 4],
    [2, 4, 1, 3, 2, 1, 3, 4],
    [3, 1, 4, 2, 4, 2, 3, 1],
    [4, 2, 3, 1, 3, 4, 2, 1],
    [2, 3, 1, 4, 1, 3, 2, 4],
    [3, 1, 4, 2, 4, 2, 1, 3],
    [4, 2, 3, 1, 3, 1, 4, 2],
    [1, 4, 2, 3, 2, 3, 1, 4]
]

def generateInitialPopulation(nrIndividuals):
    population = []
    for _ in range(nrIndividuals):
        population.append(random.sample(range(N), N))
    return population


def calculateFitness(individual):
    #calculate weight
    weight = 0
    for index in range(N):
        row = index
        column = individual[index]
        weight+=weights[row][column]
#         for j in range(index+1,N):
#             row2 = j
#             column2 = individual[j]
#             if abs(row-row2) == abs(column - column2):
#                 weight -= 1
    #check if the queens threaten each other
    
    for i in range(N):
        for j in range(i+1,N):
            if abs(individual[i]-individual[j]) == abs(i-j):
                weight-=1 #penalizing the threatening positions
    return weight

def selectParents(population, nrParents):
    #tournament selection
    parents = []
    for _ in range(nrParents):
        tournament = random.sample(population, min(5, len(population)))
        bestPermutation = []
        bestValue = 0
        for pop in tournament:
            fitness = calculateFitness(pop)
            if fitness >= bestValue:
                bestPermutation = pop
                bestValue = fitness
        
        parents.append(bestPermutation)
    return parents

def crossoverParents(parents):
    parent1,parent2 = parents
    startPos = random.randint(0,N-2)
    endPos = random.randint(startPos+1,N-1)
    #order crossover
    child1 = [-1]*N
    child2 = [-1]*N
    
    child1[startPos:endPos+1] = parent1[startPos:endPos+1]
    for i in range(N):
        if parent2[i] not in child1:
            #look for position to put parent2[i]
            for j in range(N):
                if child1[j]==-1:
                    child1[j] = parent2[i]
                    break
    
    
    child2[startPos:endPos+1] = parents[1][startPos:endPos+1]
    for i in range(N):
        if parent1[i] not in child2:
            for j in range(N):
                if child2[j]==-1:
                    child2[j] = parent1[i]
                    break
#     print(child1)
#     print(child2)
    return [child1, child2]




def mutatePermutation(permutation, mutationRate):
    mutatedPermutation = permutation.copy()
    for i in range(len(permutation)):
        if random.random() < mutationRate:
            position = random.randint(0,N-1)
            mutatedPermutation[i], mutatedPermutation[position] = mutatedPermutation[position], mutatedPermutation[i]
        else:
            mutatedPermutation[i] = mutatedPermutation[i]
    
    return mutatedPermutation
    

def geneticAlgorithm(populationSize, generations, mutationRate):
    
    #generate initial population
    population = generateInitialPopulation(populationSize)
    
    #termination clause: go through nr of generations
    for _ in range(generations):
        
        #i calculate the fitness scores while searching for the parents
        
        parents = selectParents(population, 2)
    
        offspring = [] 
        for i in range(populationSize//2):
            #generate the crossovers
            children = crossoverParents(parents)
            
            #mutate the children
            mutation1 = mutatePermutation(children[0], mutationRate)
            mutation2 = mutatePermutation(children[1], mutationRate)
            offspring.append(mutation1)
            offspring.append(mutation2)
        
        #replace population 
        population = offspring
    return population

def getBestScore(population):
    bestWeight = 0
    bestIndividual = []
    
    for index in range(len(population)):
        fitness = calculateFitness(population[index])
        if fitness > bestWeight:
            bestWeight = fitness
            bestIndividual = population[index]
    
    return  bestWeight, bestIndividual


finalPopulation = geneticAlgorithm(100, 1000,0.1)
bestWeight, bestIndividual = getBestScore(finalPopulation)

print(f'max weight: {bestWeight}')
print(f'best individual: {bestIndividual}')


```
